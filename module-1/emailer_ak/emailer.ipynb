```python
import os
from dotenv import load_dotenv
import smtplib
from email.message import EmailMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import HumanMessage, SystemMessage

load_dotenv()
```

```python
GMAIL_USER = "adityakiran.91@gmail.com"
GMAIL_APP_PASSWORD = os.environ.get("MY_GMAIL_PASS")
print(len(GMAIL_APP_PASSWORD))
```

```python
# 1. Setup Configuration
# Replace with your actual file path
RESUME_PATH = r"D:\Code\ak\career_resumes_repo\industry_2025\1-page resume\neonit\resume_ds_exp.pdf"
GMAIL_USER = "adityakiran.91@gmail.com"
# GMAIL_APP_PASSWORD = "eado pfmq vjvw lokh" # Use Google App Passwords, not your login password
GMAIL_APP_PASSWORD = os.environ.get("MY_GMAIL_PASS")

# 2. Initialize Gemini
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
)

def generate_email_content(recruiter_email, role_name):
    """Uses Gemini to draft a personalized email body."""
    
    # system_msg = """
    # "You are an expert technical recruiter and career coach. Your style is minimalist, professional, and confident. 
    # Rules: 
    # - Do not use 'I hope this finds you well.' No corporate fluff. 
    # - Always mention that the resume is attached. Max 50 words.
    # - DO NOT include a Subject line or To field"
    # """
    # user_prompt = f"""
    # Write a professional and concise cold email to a recruiter for a {role_name} position.
    # """

    system_msg = "Extract first name from email for a greeting. Output ONLY the name string."
    user_prompt = f"""Target: {recruiter_email}
    Rules:
    1. If Indian name found, return it.
    2. If generic (HR/Info), return 'Recruiter'.
    3. If unsure, return the username.
    4. One word only. No intro."""    

    messages = [
        SystemMessage(content=system_msg),
        HumanMessage(content=user_prompt)
    ]

    response = llm.invoke(messages)
    recruiter_name = response.content
        
    body = f"""
    <html>
        <body>
            Hi {recruiter_name}!<br>
            &nbsp;&nbsp;&nbsp;&nbsp;I have a very strong profile in ML/AI along with a PhD in Data Science. 
            I have over 8yrs experience building production-grade AI/ML solutions. 
            I am very excited about your recent posting for <b>{role_name}</b>. 
            It is an excellent fit for my profile. <b>And I meet all the required qualifications.</b></p>

            <p>Please review my <b>attached resume</b> and pass it on to the hiring manager.</p>

            <p>I am on an H1B visa. I have an employer who is sponsoring my visa. 
            So I am also open to working as a C2C contractor. 
            Please contact me at 312-804-9490 if you like my resume.</p>

            <p>Looking forward to hearing from you,<br>
            Aditya</p>
        </body>
    </html>
    """
    return body


def send_recruiter_email(recruiter_email, role_name):
    # Draft the body
    email_body = generate_email_content(recruiter_email, role_name)
    # print(f"\n {email_body} \n")
    
    # Create the email container
    msg = EmailMessage()
    msg['Subject'] = f"Excellent resume for {role_name}"
    msg['From'] = GMAIL_USER
    msg['To'] = recruiter_email
    # msg.set_content(email_body)
    msg.add_alternative(email_body, subtype="html")

    # Attach the PDF
    with open(RESUME_PATH, 'rb') as f:
        file_data = f.read()
        file_name = os.path.basename(RESUME_PATH)
        msg.add_attachment(
            file_data,
            maintype='application',
            subtype='pdf',
            filename=file_name
        )

    # Send via Gmail SMTP
    try:
        with smtplib.SMTP_SSL('smtp.gmail.com', 465) as smtp:
            smtp.login(GMAIL_USER, GMAIL_APP_PASSWORD)
            smtp.send_message(msg)
        print(f"Yay! Email successfully sent to {recruiter_email}")
    except Exception as e:
        print(f"Sorry! Failed to send email: {e}")
```

```python
send_recruiter_email("adityakiran.91@gmail.com", "Junior Data Scientist")
```
