```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pos = 300
ndim = 150

vec = np.zeros(shape=(ndim+5))
for i in range((ndim)//2 + 1):
    vec[2*i] = np.sin(pos/(10000**(2*i/ndim)))
    vec[2*i + 1] = np.cos(pos/(10000**(2*i/ndim)))
vec = vec[0:ndim].copy()

plt.scatter(np.arange(ndim), vec)
# plt.plot(np.arange(ndim), vec, marker='o', linestyle='-', markersize=4)
plt.show()
```

```python
# all positions
import numpy as np
import matplotlib.pyplot as plt

def get_positional_encoding(max_seq_len, d_model):
    # Create a matrix of shape (max_seq_len, d_model)
    pe = np.zeros((max_seq_len, d_model))
    
    # Calculate the positions and the dimensions
    position = np.arange(max_seq_len)[:, np.newaxis]
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
    
    # Apply sine to even indices and cosine to odd indices
    pe[:, 0::2] = np.sin(position * div_term)
    pe[:, 1::2] = np.cos(position * div_term)
    
    return pe

# Usage
ndim = 150
positions = 50
pe_matrix = get_positional_encoding(positions, ndim)

# Visualize Position 3 (like your code)
plt.scatter(np.arange(ndim), pe_matrix[3], s=10)
plt.title("PE Vector for Position 3")
plt.show()

# Visualize the whole matrix
plt.imshow(pe_matrix, cmap='RdBu', aspect='auto')
plt.xlabel("Dimension")
plt.ylabel("Position")
plt.colorbar()
plt.show()
```
