## Context

At LangChain, we aim to make it easy to build LLM applications. One type of LLM application you can build is an agent. There’s a lot of excitement around building agents because they can automate a wide range of tasks that were previously impossible. 

In practice though, it is incredibly difficult to build systems that reliably execute on these tasks. As we’ve worked with our users to put agents into production, we’ve learned that more control is often necessary. You might need an agent to always call a specific tool first or use different prompts based on its state. 

To tackle this problem, we’ve built [LangGraph](https://docs.langchain.com/oss/python/langgraph/overview) — a framework for building agent and multi-agent applications. Separate from the LangChain package, LangGraph’s core design philosophy is to help developers add better precision and control into agent workflows, suitable for the complexity of real-world systems.

### Structured output  

```python

from typing import List
from pydantic import BaseModel, Field

class PopularFunctions(BaseModel):
    method_list: List[str] = Field(description="a list of popular functions")
    pass
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
llm_model_with_struct = llm.with_structured_output(PopularFunctions)

output_obj = llm_model_with_struct.invoke("Give me the top 5 most used string methods in Python")
print(output_obj.method_list)
```

### Tool calling

```python
# Using llm.bind
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_core.messages import HumanMessage, ToolMessage

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
search_tool = TavilySearchResults(max_results=5)
llm_with_tools = llm.bind_tools([search_tool])

# ask query
# llm.invoke("What is Brad Pitt's birthday? how old is he?")
messages = [HumanMessage(content="How's the weather in Tokyo currently")]
ai_msg = llm_with_tools.invoke(messages) #ONLY decides if tool call is required. Additionally the tool must be called

print(ai_msg.tool_calls)
```

```python
# 4. if the LLM wanted to use a tool, invoke the tool and append its output to the original query
if ai_msg.tool_calls:
    messages.append(ai_msg)
    for tool_call in ai_msg.tool_calls:
        # Manually execute the tool
        print(f" LLM requested tool: {tool_call['name']}")
        result = search_tool.invoke(tool_call["args"])
        
        # Print the raw research result
        print(" Search Result:", result)
        
        # (Optional) Feed it back to the LLM for a final conversational answer
        messages.append(ToolMessage(content=str(result), tool_call_id=tool_call["id"]))
        final_response = llm_with_tools.invoke(messages)
        print("\n AI Final Answer:", final_response.content)
else:
    print("🤖 AI answered directly:", ai_msg.content)
```

### trying parallel tool calls (done by the llm)



```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
search_tool = TavilySearchResults(max_results=5)
llm_with_tools = llm.bind_tools([search_tool])

messages = [HumanMessage(content="How's the weather in Tokyo? And also give current weather in Paris")]
ai_msg = llm_with_tools.invoke(messages) #ONLY decides if tool call is required. Additionally the tool must be called

ai_msg.tool_calls
```

```python
if ai_msg.tool_calls:
    messages.append(ai_msg) # append the ai_msg output
    for tool_call in ai_msg.tool_calls:
        if tool_call:
            tool_result = search_tool.invoke(tool_call["args"])
            tool_msg = ToolMessage(content=str(tool_result), tool_call_id=tool_call["id"])
            messages.append(tool_msg) # append the tool
messages
```

```python
final_output = llm_with_tools.invoke(messages)
print(final_output.content)
```
